THE FORTUNE 1000 CSV

```python
# step 1: import library
import pandas as pd
```

```python
# step 2: import data
fortune = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Fortune1000.csv')
```

```python
fortune.head()
```

```python
fortune.info()
```

```python
fortune.describe()
```

```python
# step 3: define Y and X
fortune.columns
```

```python
Y = fortune['Employees']
X = fortune[['Rank', 'Company', 'Sector', 'Industry', 'Location', 'Revenue',
       'Profits']]
X = pd.get_dummies(X, columns=['Sector', 'Industry', 'Location'])
```

```python
# step 4:  split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=2529)
X_train = X_train.drop('Company', axis=1)
X_test = X_test.drop('Company', axis=1)
```

```python
# step 5: select  model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# step 6 : train select (fit model )
model.fit(X_train, Y_train)
```

```python
#step 7: prediction
model.predict(X_test)

```

```python
#step 8: accuracy
from sklearn.metrics import accuracy_score
Y_pred = model.predict(X_test)
```
